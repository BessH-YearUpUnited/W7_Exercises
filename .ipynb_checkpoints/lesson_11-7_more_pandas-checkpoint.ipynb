```python
for i in df.index:
    if '1-3' in df.loc[i, 'TV_Amount']:
        df.loc[i, 'TV_Amount'] = 2
    elif '3-6' in df.loc[i, 'TV_Amount']:
        df.loc[i, 'TV_Amount'] = 4.5
    elif '6-8' in df.loc[i, 'TV_Amount']:
        df.loc[i, 'TV_Amount'] = 7
    elif 'More than 8' in df.loc[i, 'TV_Amount']:
        df.loc[i, 'TV_Amount'] = 12
    elif 'Less than 1' in df.loc[i, 'TV_Amount']:
        df.loc[i, 'TV_Amount'] = .5
    else:
        df.loc[i, 'TV_Amount'] = 0
```

### Merging DataFrames

```python
inventory = pd.DataFrame({ "prod_num": [658, 2476, 3911], "quantity": [120, 54, 98]})
inventory
```

```python
sales_det = pd.DataFrame({ "prod_num": [658, 2476, 658, 2476, 3911, 3911], "price": [7.89, 5.49, 7.89, 5.49, 6.25, 6.25]})
sales_det
```

```python
sales = sales_det.groupby(['prod_num']).mean()
sales
```

```python
merged_df = pd.merge(inventory, sales, on=['prod_num'])
merged_df
```

```python
merged_df2 = pd.merge(inventory, sales_det, on=['prod_num'])
merged_df2
```

```python
merged_df2.drop_duplicates(inplace=True)
```

```python
merged_df2
```

### Example of working with joined data an a starting point

```python
df2 = pd.read_json("country_data.json")
df2
```

```python
# First let's break out the country information into a separate DataFrame, then remove duplicates

df2_a = df2.drop('Language', axis=1)
df2_a.drop_duplicates(inplace=True)
```

```python
# New DataFrame for country info:

df2_a.head(3)
```

```python
# Next let's create a new DataFrame listing languages and corresponding countries where that language is spoken

df2_b = pd.DataFrame((df2['Language'], df2['Name']))
```

```python
# Checking what our new DataFrame looks like, we'll find that it's not quite the right shape!

df2_b.head(5)
```

```python
# If we use the transpose attribute (.T) to flip the data on the column/row axis, this will look much more useful

df2_b = df2_b.T
df2_b.head(5)
```

```python
# Let's also sort this new DataFrame by language instead of by country

df2_b.sort_values('Language', inplace=True)
df2_b.head(10)
```
